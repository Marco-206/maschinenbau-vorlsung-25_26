## Aufgabe 1 ⭐ – `md5_von`

Der String wird als UTF-8 kodiert und dem MD5-Konstruktor übergeben. `.hexdigest()` liefert die 32 Zeichen lange Hex-Repräsentation.

```python
import hashlib

def md5_von(text):
    return hashlib.md5(text.encode("utf-8")).hexdigest()

print(md5_von("hallo"))
assert md5_von("hallo") == "598d4c200461b81522a3328565c25f7c"
assert len(md5_von("egal")) == 32
print("✅ md5_von korrekt.")
```

## Aufgabe 2 ⭐ – `sha256_von`

Völlig analog, nur mit `hashlib.sha256`. Der Hex-String ist 64 Zeichen lang.

```python
import hashlib

def sha256_von(text):
    return hashlib.sha256(text.encode("utf-8")).hexdigest()

print(sha256_von("test"))
assert sha256_von("test") == "9f86d081884c7d659a2feaa0c55ad015a3bf4f1b2b0b822cd15d6c15b0f00a08"
assert len(sha256_von("egal")) == 64
print("✅ sha256_von korrekt.")
```

## Aufgabe 3 ⭐ – `passwort_korrekt`

Wir hashen die Eingabe mit SHA-256 und vergleichen das Ergebnis direkt mit dem gespeicherten Hash-String. `==` liefert automatisch `True` oder `False`.

```python
import hashlib

def passwort_korrekt(eingabe, gespeicherter_sha256):
    return hashlib.sha256(eingabe.encode("utf-8")).hexdigest() == gespeicherter_sha256

gespeichert = hashlib.sha256(b"admin").hexdigest()
assert passwort_korrekt("admin", gespeichert) is True
assert passwort_korrekt("falsch", gespeichert) is False
assert passwort_korrekt("Admin", gespeichert) is False
print("✅ passwort_korrekt korrekt.")
```

## Aufgabe 4 ⭐⭐ – `knacke` (Wörterbuch-Angriff)

Wir laufen die Wortliste durch, hashen jedes Wort mit SHA-256 und vergleichen mit dem Ziel-Hash. Bei Treffer wird das Klartext-Wort zurückgegeben, andernfalls nach der Schleife `None`.

```python
import hashlib

def knacke(ziel_hash, wortliste):
    for wort in wortliste:
        if hashlib.sha256(wort.encode("utf-8")).hexdigest() == ziel_hash:
            return wort
    return None

ziel = hashlib.sha256(b"admin").hexdigest()
liste = ["1234", "test", "admin", "qwerty", "passwort"]
print("Gefundenes Passwort:", knacke(ziel, liste))

assert knacke(ziel, liste) == "admin"
assert knacke(hashlib.sha256(b"nicht_drin").hexdigest(), liste) is None
print("✅ knacke korrekt.")
```

## Aufgabe 5 ⭐⭐ – `firmware_valide`

Die Datei wird im **Binärmodus** (`"rb"`) geöffnet und komplett eingelesen. Der SHA-256-Hash der Bytes wird berechnet und mit dem erwarteten Hash verglichen. Im Produktiv-Einsatz würde man bei sehr großen Dateien blockweise lesen, für unsere 512 kB reicht das Einmal-Lesen problemlos aus.

Beachte den Pfad: Vom `loesungen/`-Ordner aus liegt `testdaten/` ein Verzeichnis höher, wir schreiben also `../testdaten/firmware_v2.3.0.bin`.

```python
import hashlib

def firmware_valide(bin_pfad, erwarteter_hash):
    with open(bin_pfad, "rb") as f:
        daten = f.read()
    tatsaechlicher_hash = hashlib.sha256(daten).hexdigest()
    return tatsaechlicher_hash == erwarteter_hash

pfad = "../testdaten/firmware_v2.3.0.bin"

# Echten Hash berechnen, um den Assert reproduzierbar zu machen
with open(pfad, "rb") as f:
    echter_hash = hashlib.sha256(f.read()).hexdigest()
print(f"SHA-256 der Firmware: {echter_hash}")

assert firmware_valide(pfad, echter_hash) is True
assert firmware_valide(pfad, "0" * 64) is False
print("✅ firmware_valide korrekt.")
```
