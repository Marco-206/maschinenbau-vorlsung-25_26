## 6. MD5 von „geheim"

```python
import hashlib
ergebnis = hashlib.md5("geheim".encode("utf-8")).hexdigest()
print(ergebnis)

erwartet = hashlib.md5(b"geheim").hexdigest()
assert ergebnis == erwartet
print("✅ Richtig!")
```

## 8. Passwort-Check

Die Funktion hasht die Eingabe mit SHA-256 und vergleicht das Ergebnis mit dem gespeicherten Hex-Wert.

```python
import hashlib

def meine_pruefung(eingabe, hash_wert):
    h = hashlib.sha256(eingabe.encode("utf-8")).hexdigest()
    return h == hash_wert

h = hashlib.sha256(b"robot").hexdigest()
assert meine_pruefung("robot", h) is True
assert meine_pruefung("falsch", h) is False
print("✅ Richtig!")
```

## 9. SHA-256-Hash-Länge

```python
import hashlib

laenge_sha256 = len(hashlib.sha256(b"irgendwas").hexdigest())
print(laenge_sha256)

assert laenge_sha256 == 64
print("✅ Richtig – 64 Hex-Zeichen = 256 Bit.")
```

## 10. Mini-Wörterbuch-Angriff

```python
import hashlib

ziel_hash = hashlib.sha256(b"admin").hexdigest()
wortliste = ["test", "1234", "admin", "passwort"]

knackpasswort = None
for kandidat in wortliste:
    if hashlib.sha256(kandidat.encode("utf-8")).hexdigest() == ziel_hash:
        knackpasswort = kandidat
        break

print("Geknacktes Passwort:", knackpasswort)
assert knackpasswort == "admin"
print("✅ Richtig – Wörterbuch-Angriff erfolgreich.")
```
